```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

import statsmodels.api as sm
```

```python
weather12mon_df=pd.DataFrame(columns = ["cloud", "humid", "precip", "tempature","housing"])
for x in range(12):
    weather12mon_df.loc[x]=[cloud_data[x],humid_data[x],precip_data[x],temp_data[x],housing_data[x]]

weather12mon_df  

# y=weather12mon_df.iloc[:,4].values
# y

```

```python
reg = LinearRegression()
Xs = weather12mon_df[["cloud","humid","precip","tempature"]]
y = weather12mon_df.iloc[:,4].values
reg.fit(Xs, y)
```

```python

print(" Y = "+str(round(reg.intercept_,2))+" + "+str(round(reg.coef_[0],2))+"*cloud " +str(round(reg.coef_[1],2))+"*humid "+str(round(reg.coef_[2],2))+"*precip "+" + "+str(round(reg.coef_[3],2))+"*temperature")
```
