```python
#Dependencies
import pandas as pd
import numpy as np
```

```python
#Read in cleaned data set
all_data = pd.read_csv("Resources/cleaned_data.csv")
all_data.head()
```

```python
#Merge in region information

#Read in region data
region_data = pd.read_csv("Resources/Regions.csv")
region_data.head()
```

```python
all_region = pd.merge(all_data, region_data, on="State")
all_region.head()
```

```python
#Merge in population

#Read in population data
pop_data = pd.read_csv("Resources/cities_pop.csv")
pop_data.head()
```

```python
full_data = pd.merge(all_region, pop_data, on=["City", "State"])
full_data.head()
```

```python
#Distribution of Home Sale Counts by month:  four regions
region4_grp = full_data.groupby(["Region4"])

months = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]

for month in months:
    Total_month = all_region[f"{month}_Avg"].sum()
    month_sums = region4_grp[f"{month}_Avg"].sum()
    month_pct = month_sums/Total_month

    month_pct_df = pd.DataFrame({f"Home Sales {month}": month_sums, "Percent": month_pct})
    month_pct_df["Percent"] = month_pct_df["Percent"].map("{:.2%}".format)
    print(month_pct_df)
```

```python
#Distribution of Home Sale Counts by month:  nine regions
region9_grp = full_data.groupby(["Region9"])

months = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]

for month in months:
    Total_month = all_region[f"{month}_Avg"].sum()
    month_sums = region9_grp[f"{month}_Avg"].sum()
    month_pct = month_sums/Total_month

    month_pct_df = pd.DataFrame({f"Home Sales {month}": month_sums, "Percent": month_pct})
    month_pct_df["Percent"] = month_pct_df["Percent"].map("{:.2%}".format)
    print(month_pct_df)
```

```python
#Population distribution by region:  4 regions
Total_Pop = full_data["Population"].sum()
region_sums = region4_grp["Population"].sum()
region_pct = region_sums/Total_Pop

Pop4_pct_df = pd.DataFrame({"Region Population": region_sums, "Percent": region_pct})
Pop4_pct_df["Percent"] = Pop4_pct_df["Percent"].map("{:.2%}".format)
Pop4_pct_df
```

```python
#Population distribution by region:  9 regions
Total_Pop = full_data["Population"].sum()
region_sums = region9_grp["Population"].sum()
region_pct = region_sums/Total_Pop

Pop9_pct_df = pd.DataFrame({"Region Population": region_sums, "Percent": region_pct})
Pop9_pct_df["Percent"] = Pop9_pct_df["Percent"].map("{:.2%}".format)
Pop9_pct_df
```

```python
#Create output file of data
output_data_file = "resources/final_data.csv"
full_data.to_csv(output_data_file, index=False, header=True)
```
