```python
#Dependencies
import pandas as pd
```

```python
#Combine all data sets from API calls together into one data frame

#Read in data
Group0 = pd.read_csv("Resources/P1_cities_WD_G0.csv")
Group1 = pd.read_csv("Resources/P1_cities_WD_G1.csv")
Group2 = pd.read_csv("Resources/P1_cities_WD_G2.csv")
Group3 = pd.read_csv("Resources/P1_cities_WD_G3.csv")
Group4 = pd.read_csv("Resources/P1_cities_WD_G4.csv")
Group5 = pd.read_csv("Resources/P1_cities_WD_G5.csv")
Group6 = pd.read_csv("Resources/P1_cities_WD_G6.csv")
Group7 = pd.read_csv("Resources/P1_cities_WD_G7.csv")
Group8 = pd.read_csv("Resources/P1_cities_WD_G8.csv")
Group9 = pd.read_csv("Resources/P1_cities_WD_G9.csv")
Group10 = pd.read_csv("Resources/P1_cities_WD_G10.csv")

all_data = pd.concat([Group0, Group1, Group2, Group3, Group4, Group5, Group6, Group7,
                     Group8, Group9, Group10], axis=0).reset_index(drop=True)
all_data.head()
```

```python
#Create output file of entire (uncleaned) data set
output_data_file = "resources/all_data_uncleaned.csv"
all_data.to_csv(output_data_file, index=False, header=True)
```

```python
#Missing values

#if we decide to drop all rows with any NaN values
drop_null = all_data.dropna().reset_index(drop=True)
len(drop_null.index)
```

```python
#Missing values

#if we decide to keep - then replace with zeros
all_new = all_data.fillna(0)
all_new.head(20)
```

```python
#Keep only records with at least one full year of data

#Flag rows that will need to be dropped
all_new["flag17"] = ""
all_new["flag18"] = ""
all_new["flag19"] = ""

years = ["2017", "2018", "2019"]

for index, row in all_new.iterrows(): 
    
    for year in years:
        
        count = 0
        
        if row[f"{year}-01"] > 0:
            count = count + 1
        if row[f"{year}-02"] > 0:
            count = count + 1
        if row[f"{year}-03"] > 0:
            count = count + 1
        if row[f"{year}-04"] > 0:
            count = count + 1
        if row[f"{year}-05"] > 0:
            count = count + 1
        if row[f"{year}-06"] > 0:
            count = count + 1
        if row[f"{year}-07"] > 0:
            count = count + 1
        if row[f"{year}-08"] > 0:
            count = count + 1
        if row[f"{year}-09"] > 0:
            count = count + 1
        if row[f"{year}-10"] > 0:
            count = count + 1
        if row[f"{year}-11"] > 0:
            count = count + 1
        if row[f"{year}-12"] > 0:
            count = count + 1
            
        if year == "2017":
            all_new.loc[index, "flag17"] = count
        elif year == "2018":
            all_new.loc[index, "flag18"] = count
        elif year == "2019":
            all_new.loc[index, "flag19"] = count

all_new.head()
```

```python
#Drop rows without at least one full year of Zillow data
all_new = all_new.loc[(all_new['flag17'] == 12) | (all_new['flag18'] == 12) | (all_new['flag19'] == 12)]
```

```python
#Summarize Zillow data:  averages for each month across the three year period

#new columns of data
all_new["01_Avg"] = ""
all_new["02_Avg"] = ""
all_new["03_Avg"] = ""
all_new["04_Avg"] = ""
all_new["05_Avg"] = ""
all_new["06_Avg"] = ""
all_new["07_Avg"] = ""
all_new["08_Avg"] = ""
all_new["09_Avg"] = ""
all_new["10_Avg"] = ""
all_new["11_Avg"] = ""
all_new["12_Avg"] = ""

months = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]

for index, row in all_new.iterrows():
    
    for month in months:
        
        count = 0
        sums = 0
    
        if row[f"2017-{month}"] > 0:
            count = count + 1
            sums = sums + row[f"2017-{month}"]
        if row[f"2018-{month}"] > 0:
            count = count + 1
            sums = sums + row[f"2018-{month}"]
        if row[f"2019-{month}"] > 0:
            count = count + 1
            sums = sums + row[f"2019-{month}"]
        
        if count > 0:
            avg = round(sums/count)
        else:
            avg = 0 
        
        all_new.loc[index, (f"{month}_Avg")] = avg

    
all_new.head()
```

```python
#Drop old Zillow columns
clean_data = all_new[["City", "State", "Latitude", "Longitude",
                      "01_Avg", "02_Avg", "03_Avg", "04_Avg", "05_Avg", "06_Avg",
                      "07_Avg", "08_Avg", "09_Avg", "10_Avg", "11_Avg", "12_Avg",
                      "Jan_temp", "Jan_precip", "Jan_humid", "Jan_cloud",
                      "Feb_temp", "Feb_precip", "Feb_humid", "Feb_cloud",
                      "Mar_temp", "Mar_precip", "Mar_humid", "Mar_cloud",
                      "Apr_temp", "Apr_precip", "Apr_humid", "Apr_cloud",
                      "May_temp", "May_precip", "May_humid", "May_cloud",                      
                      "Jun_temp", "Jun_precip", "Jun_humid", "Jun_cloud",                      
                      "Jul_temp", "Jul_precip", "Jul_humid", "Jul_cloud",                      
                      "Aug_temp", "Aug_precip", "Aug_humid", "Aug_cloud",                      
                      "Sep_temp", "Sep_precip", "Sep_humid", "Sep_cloud",                     
                      "Oct_temp", "Oct_precip", "Oct_humid", "Oct_cloud",
                      "Nov_temp", "Nov_precip", "Nov_humid", "Nov_cloud",                     
                      "Dec_temp", "Dec_precip", "Dec_humid", "Dec_cloud"                     
                     ]]
clean_data.head()
```

```python
#Output cleaned data
output_data_file = "resources/cleaned_data.csv"
clean_data.to_csv(output_data_file, index=False, header=True)
```
